## In this exercise we'll be scraping the poll data from RCP site and analyse it!

First we'll scrap the data

```python
def get_poll_xml(poll_id):
    xml = requests.get("http://charts.realclearpolitics.com/charts/" + poll_id + ".xml").text
    return xml
#get_poll_xml("1171")
xml = get_poll_xml("1171")
```

```python
def _strip(s):
    """This function removes non-letter characters from a word
    
    for example _strip('Hi there!') == 'Hi there'
    """
    return re.sub(r'[\W_]+', '', s)
```

```python
#instead of pattern library I'm using beautiful soup as pattern doesn't support python 3
def plot_colors(xml):
    dom = bs4.BeautifulSoup(xml, "lxml")
    result = {}
    for graph in dom.find_all('graph'): 
            title = _strip(graph['title'])
            result[title] = graph['color']
    return result        
#print (result)
```

```python
#cleaning the data is really important, make the array sizes equal and then convert it into a panda
#dataframe
def rcp_poll_data(xml):
    dom = bs4.BeautifulSoup(xml, "lxml")
    values = dom.find("series").find_all("value")
    value_list = []
    for value in values:
        value_list.append(value.text)
        #print (value_list)    
    presidents = {}
    graph = dom.find_all("graph")
    result = {}
    result['Date'] = pd.to_datetime(value_list, errors = "ignore")
    for g in graph:
        name = g["title"]
        data = [float(n.text) if n.text else np.nan for n in g]
        result[name] = data
        
    result = pd.DataFrame(result)
    
    #    result.Obama = result.Obama.astype(float)
    #    result.Romney = result.Romney.astype(float)
    return result
```

```python
rcp_poll_data(xml).head()
```

```python
def poll_plot(poll_id):
    xml = get_poll_xml(poll_id)
    #returns a panda dataframe...
    data = rcp_poll_data(xml)
    colors = plot_colors(xml)
    
    #remove characters like apostrophes
    data = data.rename(columns = {c: _strip(c) for c in data.columns})
    
    #normalize poll numbers so they add to 100%    
    norm = data[list(colors.keys())].sum(axis = 1) / 100
    for c in colors.keys():
        data[c] /= norm
    
    for label, color in colors.items():
        plt.plot(data.Date, data[label], color = color, label = label)
    plt.xticks(rotation=70)
    plt.legend(loc='best')
    plt.xlabel("Date")
    plt.ylabel("Normalized Poll Percentage") 
rcp_poll_data(xml).dtypes
```

### Unfortunately, these data don't have any error bars. If a candidate leads by 10% in the RCP average, is she a shoo-in to win? Or is this number too close to call? Does a 10% poll lead mean more 1 day before a race than it does 1 week before? Without error estimates, these questions are impossible to answer.

To get a sense of how accurate the RCP polls are, you will gather data from many previous Governor races, where the outcome is known.

This url has links to many governer races.

http://www.realclearpolitics.com/epolls/2010/governor/2010_elections_governor_map.html

Each link to a governor race has the following URL pattern:

http://www.realclearpolitics.com/epolls/[YEAR]/governor/[STATE]/[TITLE]-[ID].html

```python
text = requests.get("http://www.realclearpolitics.com/epolls/2010/governor/2010_elections_governor_map.html").text
soup = bs4.BeautifulSoup(text, "html.parser")
```

```python
def find_governor_races():
    clear_fix = soup.find("div", {'id':'states'})
    url_links = clear_fix.find_all("a")
    #list-comprehension, add only those elements which have /epolls in starting
    list_urls = [l["href"] for l in url_links if "/epolls" in l["href"] ]
    result = []
    for l in list_urls:
        result.append("http://www.realclearpolitics.com" + l)
    #eliminate duplicates!    
    result = list(set(result))
    return result
```

```python
#sample output
urls = find_governor_races()
urls
```

```python
def race_result(url):
    req = requests.get(url).text
    soup = bs4.BeautifulSoup(req, "html.parser")    
    data = soup.find('div', {'id':'polling-data-rcp'})
    table = data.table
    trs = table.find_all('tr')
    first_row = trs[0].find_all('th')[3:-1]
    second_row = trs[1].find_all('td')[3:-1]
    #these are the headers
    labels = [t.text for t in first_row]

    #these are the results 
    results = [float(t.text) for t in second_row]
    tot = sum(results) / 100
    #remove the data or useless string ahead of the names
    labels = [str(l).split('(')[0].strip() for l in labels]
    for l in labels:
        l = _strip(l)
    fin_result = {_strip(l) : r / tot for l, r in zip(labels, results)}
    return fin_result
race_result(urls[5])
```

```python
def id_from_url(url):
    """Given a URL, look up the RCP identifier number"""
    return url.split('-')[-1].split('.html')[0]

def plot_race(url):
    """Make a plot summarizing a senate race
    
    Overplots the actual race results as dashed horizontal lines
    """
    #hey, thanks again for these functions!
    poll_id = id_from_url(url)
    xml = get_poll_xml(str(poll_id))    
    colors = plot_colors(xml)

    if len(colors) == 0:
        return
    
    #really, you shouldn't have
    result = race_result(url)
    
    poll_plot(poll_id)
    plt.xlabel("Date")
    plt.ylabel("Polling Percentage")
    #the below code shows the final results in dotted lines!
    for r in result:
        plt.axhline(result[r], color=colors[_strip(r)], alpha=0.6, ls='--')
```

```python
for u in urls:
    plot_race(u)
    plt.show()
```

Well these graphs tell us a very interesting thing i.e. the actual election outcomes are in 5% range of the the predictions in most of the cases. In about 20% of cases, the eventual winner trails at some point during the last month of polling -- examples include the Brady/Quinn Illinois race, the Foley/Malloy Connecticut race, the Scott/Sink Florida Race, the Kitzhaber/Dudley Oregon Race, the Corzine/Christie New Jersey race, and the Caprio/Robitaille/Chafee Rhode Island Race.

## Now the analysis! We'll be using bootstrap tecnique to do the analysis

`error_data` function computes a new DataFrame with information about polling errors.

```python
def party_from_color(color):
    if color in ['#0000CC', '#3B5998']:
        return 'democrat'
    if color in ['#FF0000', '#D30015']:
        return 'republican'
    return 'other'


def error_data(url):
    """
    Given a Governor race URL, download the poll data and race result,
    and construct a DataFrame with the following columns:
    
    candidate: Name of the candidate
    forecast_length: Number of days before the election
    percentage: The percent of poll votes a candidate has.
                Normalized to that the canddidate percentages add to 100%
    error: Difference between percentage and actual race reulst
    party: Political party of the candidate
    
    The data are resampled as necessary, to provide one data point per day
    """
    #extract the id
    poll_id = id_from_url(url)
    #get the xml data
    xml = get_poll_xml(poll_id)
    
    #get the colors from the xml data
    colors = plot_colors(xml)
    if len(colors) == 0:
        return pd.DataFrame()
    #get the dataframe from the xml data
    df = rcp_poll_data(xml)
    #get the final race result from the url
    result = race_result(url)
    print (result)
    #remove non-letter characters from columns
    #rename accepts a dictionary representing names of the columns 
    df = df.rename(columns={c: _strip(c) for c in df.columns})
    #do the same for the final result dictionary 
    #####CAUGHT AN IMP ERROR HERE#######
    #you cannot change the keys in a dictionary as it'll alter its size...
    for k, v in result.items():
        result[_strip(k)] = v 
    
    #get the candidate names from the dataframe
    candidates = [c for c in df.columns if c is not 'Date']
        
    #change the index of the dataframe to the dates turn into a timeseries...
    df.index = df.Date

    
    #...so that we can resample at regular, daily intervals
    df = df.resample('D').mean()
    df = df.dropna()
    #print (df.Date)    
    #compute forecast length in days
    #(assuming that last forecast happens on the day of the election, for simplicity)
    forecast_length = (df.index.max() - df.index).values
    forecast_length = forecast_length / np.timedelta64(1, 'D')  # convert to number of days
    #print (forecast_length, df.index.max(), type(df.index))

    
    #compute forecast error
    errors = {}
    #comput it from 100
    normalized = {}
    poll_lead = {}
    
    for c in candidates:
        #turn raw percentage into percentage of poll votes
        #in numpy axis = 1 is rows 
        corr = df[c].values / df[candidates].sum(axis=1).values * 100.
        err = corr - result[c]
        
        normalized[c] = corr
        errors[c] = err
        
    n = forecast_length.size
    
    result = {}
    result['percentage'] = np.hstack(normalized[c] for c in candidates)
    result['error'] = np.hstack(errors[c] for c in candidates)
    result['candidate'] = np.hstack(np.repeat(c, n) for c in candidates)
    result['party'] = np.hstack(np.repeat(party_from_color(colors[_strip(c)]), n) for c in candidates)
    result['forecast_length'] = np.hstack(forecast_length for _ in candidates)
    result = pd.DataFrame(result)
    return result    
```

```python
def all_error_data():
    data = [error_data(race_page) for race_page in urls]
    return pd.concat(data, ignore_index=True)
errors = all_error_data()
```

## Bootstrap analysis

If I have the final data of an election and now I'd like to predict in the further election who'll win the race I cannot get a lot of samples to predict, as it is a one-off event. So what I do is do bootstrapping to produce random samples which are trials and then predict who'll win in the further scenario!

```python
def bootstrap(c1, c2, nsample):
    #first we'll normalise the final data
    tot = c1 + c2
    c1 = 100. * c1 / tot
    c2 = 100. * c2 / tot
    
    rand = np.random.randint(0, errors.shape[0], nsample)
    values = np.array([errors.error.iloc[i] for i in rand])
    
    c1_new = c1 + values
    c2_new = c2 - values
    
    p1 = (c1_new > c2_new).mean()
    p2 = 1 - p1
    return p1, p2
```

```python
nsample = 10000
mcauliffe, cuccinelli = 43.0, 39.0

pm, pc = bootstrap(mcauliffe, cuccinelli, nsample=nsample)
print ("Virginia Race")
print ("-------------------------")
print ("P(McAuliffe wins)  = %0.8f" % pm)
print ("P(Cuccinelli wins) = %0.8f" % pc)

#new jersey
print ("\n\n")
print ("New Jersey Race")
print ("-----------------------")
christie, buono = 55.4, 31.8
pc, pb = bootstrap(christie, buono, nsample=nsample)
print ("P(Christie wins) = %0.8f" % pc)
print ("P(Buono wins)    = %0.8f" % pb)
```

The Virginia race is currently fairly close. Our bootstrap simulations assume that historical RCP poll accuracies are representative of the uncertainty in the current polls. In 10,000 of these simulations, McCauliffe won 70% of the time.

The New Jersey race, on the other hand, is less close. In 10,000 simulations, a candidate enjoying a lead as big as Chris Christie wins about 98% of the time
